```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.random.randint(0, 100, size=100)
```

```python
y_p = 2 * x + 8
```

```python
sns.scatterplot(x, y_p)
```

```python
y = y_p + (np.random.random(100)-0.5) * 30
```

```python
sns.scatterplot(x, y)
```

```python
ax = sns.scatterplot(x=x, y=y, label='prediction')
sns.scatterplot(x=x, y=y_p, ax=ax, label='actuall')
```
